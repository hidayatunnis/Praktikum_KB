## Data Cleaning

```python
dt_rusak = pd.read_csv("diab_broke.csv")
dt_rusak 
```

### Mengecek apakah tiap kolom ada yang null

```python
dt_rusak.isna()
```

### Summary

```python
dt_rusak.isna().sum()
```

### Null Cleaning

```python
from sklearn.impute import SimpleImputer
```

### Mean

```python
imputer_mean = SimpleImputer(strategy = 'mean')
```

### Median

```python
imputer_median = SimpleImputer(strategy = 'median')
```

### Modus

```python
imputer_modus = SimpleImputer(strategy = 'most_frequent')
```

### Copy dataset

```python
dt_clean = dt_rusak.copy()
```

```python
dt_clean['Pregnancies'] = imputer_mean.fit_transform(dt_clean[['Pregnancies']])
dt_clean['Glucose'] = imputer_mean.fit_transform(dt_clean[['Glucose']])
dt_clean['SkinThickness'] = imputer_median.fit_transform(dt_clean[['SkinThickness']])
dt_clean['Insulin'] = imputer_median.fit_transform(dt_clean[['Insulin']])
dt_clean['BMI'] = imputer_modus.fit_transform(dt_clean[['BMI']])
dt_clean['DiabetesPedigreeFunction'] = imputer_mean.fit_transform(dt_clean[['DiabetesPedigreeFunction']])
dt_clean['Age'] = imputer_median.fit_transform(dt_clean[['Age']])
dt_clean['Outcome'] = imputer_modus.fit_transform(dt_clean[['Outcome']])
```

```python
dt_clean.tail(10)
```

### Duplicate Cleaning

```python
dt_clean.duplicated()
```

### Jumlah data yang duplicate

```python
dt_clean.duplicated().sum()
```

### Menghapus data yang duplicate dan menampilkan kembali summary dari data yang duplicate

```python
dt_clean.drop_duplicates(inplace=True)
dt_clean.duplicated().sum()
```
